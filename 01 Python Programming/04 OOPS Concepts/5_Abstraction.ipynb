ABSTRACTION

```python
import abc
class pwskills:  # Abstract class for the student details
  
  @abc.abstractmethod
  def student_details(self):
    pass
  
  @abc.abstractmethod
  def student_assignment(self):
    pass
  
  @abc.abstractmethod
  def student_marks(self):
    pass
```

```python
class data_science(pwskills):  # Inherited class from the abstract class
  
  def student_details(self):
    return "It will return a details of data science masters"

  def student_assignment(self):
    return "It will return a assignment of data science masters"
```

```python
class web_development(pwskills):  # Inherited class from the abstract class
  
  def student_details(self):
    return "It will return a details of web development masters"

  def student_assignment(self):
    return "It will return a assignment of web development masters"
```

```python
ds = data_science()
ds.student_details()
```

```python
wb = web_development()
wb.student_details()
```

```python
wb.student_assignment()
```
